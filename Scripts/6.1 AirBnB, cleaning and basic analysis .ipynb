# Column data type consistency 



```python
# Column list with columns to convert to date type

columns_to_date = (['review_date', 'Host Since', 'First Review', 'Last Review'])
```

```python
for columns in columns_to_date:
    df[columns] = df[columns].astype('datetime64[ns]')
```

```python
# Column list with columns to remove decimals 

columns_to_nodec = (['Review ID', 'Reviewer ID', 'Listing ID', 'Host ID'])
```

```python
for columns in columns_to_nodec:
    df[columns] = df[columns].astype(float).astype(int)
```

```python
# Column list with columns to str

columns_to_str = (['Review ID', 'Reviewer ID', 'Listing ID', 'Host ID', 'Host Response Time', 'neighbourhood', 'Neighborhood Group', 'City', 'Country', 'Property Type', 'Room Type', 'Postal Code', 'Comments'])
```

```python
for columns in columns_to_str:
    df[columns] = df[columns].astype('string')
```

```python
# Price Column to string  

df['Price'] = df['Price'].astype(str)
```

```python
#Delete all non numeric values and convert to float 

df['Price'] = df['Price'].apply(lambda x: re.sub(r'[^0-9.]', '', x)).str.replace(',', '.').astype(float)
```

```python
# Look for alphabetic values in Postal Code

alphabetical_rows = df[df['Postal Code'].notna() & df['Postal Code'].str.contains(r'[a-zA-Z]', na=False)]
```

```python
# Display all rows

pd.set_option('display.max_rows', None)
```

```python
alphabetical_rows.head(300)
```

```python
# Postal Code consistency
#Dictionary to impute inconsistent values 

Postal_Code_alphabetic = {'10997 Kreuzberg': '10997', '10777 Berlin': '10777', '13187 Pankow' : '13187', '13359 Berlin': '13359'}
```

```python
# Replace inconsistent values with dictionary

df['Postal Code'] = df['Postal Code'].replace(Postal_Code_alphabetic)
```

```python
# create list with columns  
columns_bool = (['Is Superhost', 'Instant Bookable', 'Business Travel Ready'])
```

```python
#use list of columns to map values
for column in columns_bool:
    df[column] = df[column].map({'f': False, 't': True})
```

# Encoding 

```python
# create list of coumns
columns_to_encode = ['Listing Name', 'neighbourhood', 'Neighborhood Group', 'Comments']
```

```python
# encode columns with german text and ä,ö,ü characters]
#encode to utf-8 and decode to utf-8 after
for column in columns_to_encode:
    df[column] = df[column].apply(lambda x: str(x).encode('utf-8').decode('utf-8'))
```
